**Merging GDP with the existing Dataset**

```python
import pandas as pd
import numpy as np
```

```python
og_df = pd.read_csv('Air Traffic New.csv')
og_df.head()
```

```python
og_df.set_index('Date')
```

```python
new_df = pd.read_csv('gdp.csv')
new_df.head() 
```

```python
combined_data = pd.merge(og_df,new_df,on='Date')
combined_data
```

```python
combined_data.to_csv('new_data.csv',index=False)
```
